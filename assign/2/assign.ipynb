## Άσκηση 1
Γράψτε μια συνάρτηση με το όνομα `myfunction1`, η οποία δέχεται ένα σημείο $P_1=[x_1,y_1,z_1]$ στον τρισδιάστατο χώρο.
Στην συνέχεια ορίζει άλλα δύο διανύσματα:

$$P_2=[a+1,b+1,c+1]$$

$$P_3=[a-2,b-2,c+5]$$

όπου a,b,c τα 3 τελευταία ψηφία του αριθμού μητρώου σας (c=τελευταίο).

Τέλος υπολογίζει τους συντελεστές του επιπέδου που ορίζεται από τα τρία σημεία στην μορφή:

$$c_x x+c_y y+ c_z z=1$$

και τους επιστρέφει σε ένα μονοδιάστατο πίνακα $C_f=[c_x,c_y,c_z]$ της numpy.

Ελέγξτε τα αποτελέσματα της συνάρτησης για:

- $P_1=[10,0,0]$
- $P_1=[0,10,0]$


````python
# Λύση της άσκησης 1
# - Σε αυτό το κελί γράψτε μόνο τις ζητούμενες συναρτήσεις χωρίς print και input.

import numpy as np

def myfunction1(P1: np.ndarray) -> np.ndarray:
    # Τρία τελευταία ψηφία του μητρώου
    a, b, c = 0, 2, 6

    # Ορισμός των άλλων δύο σημείων
    P2 = np.array([a+1, b+1, c+1])  # [1, 3, 7]
    P3 = np.array([a-2, b-2, c+5])  # [-2, 0, 11]

    # Διανύσματα του επιπέδου
    v1 = P2 - P1
    v2 = P3 - P1

    # Κανονικό διάνυσμα (normal vector)
    n = np.cross(v1, v2)

    # Συντελεστές επιπέδου cx*x + cy*y + cz*z = 1
    d = np.dot(n, P1)
    Cf = n / d

    return Cf

````

````python
# Πρόχειρο άσκησης 1
# - Σε αυτό το κελί μπορείτε να τυπώσετε μεταβλητές και να κάνετε γραφικές παραστάσεις για επαλήθευση. Δεν λαμβάνεται υπόψη στην βαθμολόγηση.

print(myfunction1(np.array([10.,0.,0.])), myfunction1(np.array([0.,10.,0.])))
````

## Άσκηση 2
Γράψτε μια συνάρτηση `myfunction2` που επιλύει το γραμμικό σύστημα:

$$
\begin{bmatrix}
20 & -5 & 0 & -5 &\\
-5 & 20 & -5 & 0\\
0 & -5 & 20 & -5\\
-5 & 0 & -5 & 20
\end{bmatrix}
\cdot
\begin{bmatrix} x_1\\ x_2\\ x_3 \\ x_4 \end{bmatrix}
=
\begin{bmatrix} a+1\\ b+2\\ c+3 \\ 4 \end{bmatrix}
$$
όπου a,b,c τα 3 τελευταία ψηφία του αριθμού μητρώου σας (c=τελευταίο).

Χρησιμοποιήστε την μέθοδο Gauss-Seidel σταματώντας όταν το μέτρο του απόλυτου σφάλματος της τρέχουσας επανάληψης σε σχέση με την προηγούμενη είναι μικρότερο από 2e-3 για όλα τα στοιχεία ή όταν το πλήθος των επαναλήψεων φτάσει τις 100. Η συνάρτηση θα δέχεται ως όρισμα ένα διάνυσμα $x_{init}$ με τις αρχικές τιμές και θα επιστρέφει ένα διάνυσμα 4 στοιχείων $[x_1,x_2,x_3,x_4]$ με τα τελικά αποτελέσματα.

Ελέγξτε τα αποτελέσματα της συνάρτησης για $x_{init}=[0.,0.,0.,0.]$ και $x_{init}=[100.,100.,100.,100.]$.

```
Υπόδειξη: Χρησιμοποιήστε την μέθοδο .copy(), όταν θέλετε να αντιγράψετε τις τιμές σε διαφορετικό χώρο μνήμης και όχι να δημιουργήσετε ένα δείκτη σε υπάρχοντα πίνακα.
```
```
Ερώτηση για τις εξετάσεις: Ποιο διάνυσμα αρχικών τιμών δινει πιο γρήγορη σύγκλιση και γιατί;
```

````python
# Λύση της άσκησης 2
# - Σε αυτό το κελί γράψτε μόνο τις ζητούμενες συναρτήσεις χωρίς print και input.

import numpy as np

def myfunction2(xinit: np.ndarray) -> np.ndarray:
    # Τρία τελευταία ψηφία του μητρώου
    a, b, c = 0, 2, 6

    # Πίνακας συστήματος
    A = np.array([[20, -5, -5, -5],
                  [-5, 20, -5, -5],
                  [-5, -5, 20, -5],
                  [-5, -5, -5, 20]], dtype=float)

    # Δεξιά πλευρά
    B = np.array([a+1, b+2, c+3, 4], dtype=float)

    # Αρχική εκτίμηση
    x = xinit.copy()

    tol = 2e-3
    max_iter = 100

    for it in range(max_iter):
        x_old = x.copy()

        # Gauss-Seidel update
        x[0] = (B[0] - (-5*x[1] + -5*x[2] + -5*x[3])) / 20
        x[1] = (B[1] - (-5*x[0] + -5*x[2] + -5*x[3])) / 20
        x[2] = (B[2] - (-5*x[0] + -5*x[1] + -5*x[3])) / 20
        x[3] = (B[3] - (-5*x[0] + -5*x[1] + -5*x[2])) / 20

        # Έλεγχος σύγκλισης
        if np.all(np.abs(x - x_old) < tol):
            break

    return x
````

````python
# Πρόχειρο άσκησης 2
# - Σε αυτό το κελί μπορείτε να τυπώσετε μεταβλητές και να κάνετε γραφικές παραστάσεις για επαλήθευση. Δεν λαμβάνεται υπόψη στην βαθμολόγηση.

print(myfunction2(np.array([0., 0., 0., 0.])),myfunction2(np.array([100., 100., 100., 100.])))
````
